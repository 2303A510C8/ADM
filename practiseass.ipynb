```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("scores.csv")
data.head()
```

```python
print(data.shape)
print(data.sample(5))
print(data.describe())
```

```python
plt.figure(figsize=(16, 5))
plt.subplot(1, 3, 1)
sns.histplot(data['Emotional_Quotient'])

plt.subplot(1, 3, 2)
sns.histplot(data['Intelligence_Quotient'])

plt.subplot(1, 3, 3)
sns.histplot(data['Marks'])

plt.show()
```

```python
z_score_upper_limit = data['Emotional_Quotient'].mean() + 3 * data['Emotional_Quotient'].std()
z_score_lower_limit = data['Emotional_Quotient'].mean() - 3 * data['Emotional_Quotient'].std()
print(z_score_upper_limit)
print(z_score_lower_limit)
```

```python
data['Emotional_Quotient'] = np.where(data['Emotional_Quotient'] > z_score_upper_limit, z_score_upper_limit, data['Emotional_Quotient'])
data['Emotional_Quotient'] = np.where(data['Emotional_Quotient'] < z_score_lower_limit, z_score_lower_limit, data['Emotional_Quotient'])
data.describe()
```

```python
sns.boxplot(data['Emotional_Quotient'])
data['Intelligence_Quotient'].skew()
sns.boxplot(data['Intelligence_Quotient'])

percentile25 = data['Intelligence_Quotient'].quantile(0.25)
percentile75 = data['Intelligence_Quotient'].quantile(0.75)
print(percentile25)
print(percentile75)
```

```python
iqr = percentile75 - percentile25
print(iqr)

iqr_upper_limit = percentile75 + 1.5 * iqr
iqr_lower_limit = percentile25 - 1.5 * iqr
print(iqr_upper_limit)
print(iqr_lower_limit)
```
